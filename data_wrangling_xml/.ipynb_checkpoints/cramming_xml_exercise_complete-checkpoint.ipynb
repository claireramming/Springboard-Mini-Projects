## XML example

+ for details about tree traversal and iterators, see https://docs.python.org/2.7/library/xml.etree.elementtree.html

```python
document_tree = ET.parse( './data/mondial_database_less.xml' )
```

```python
# print names of all countries and their cities
for element in document_tree.iterfind('country'):
    print('* ' + element.find('name').text + ':'), 
    print(capitals_string = '')
    for subelement in element.getiterator('city'):
        capitals_string += subelement.find('name').text + ', '
    print(capitals_string[:-2])
```

## XML exercise

Using data in 'data/mondial_database.xml', the examples above, and refering to https://docs.python.org/2.7/library/xml.etree.elementtree.html, find

1. 10 countries with the lowest infant mortality rates
2. 10 cities with the largest population
3. 10 ethnic groups with the largest overall populations (sum of best/latest estimates over all countries)
4. name and country of a) longest river, b) largest lake and c) airport at highest elevation

```python
document = ET.parse( './data/mondial_database.xml' )
```

```python
#see what's in our document
for elem in document.iter():
    print(elem.tag), print(elem.attrib)
```

```python
#1. 10 Countries with Lowest Infant Mortality rates

#initialize dictionary
im_dict = {}

# Loop to find Country name and Infant Mortality rate
for element in document.iterfind('country'):
    country = element.find('name')
    infant = element.find('infant_mortality')
    if (country != None) : 
        if(infant != None) :
            im_dict[country.text] = float(infant.text)
        else :
            im_dict[country.text] = np.nan
            
#turn it into a dataframe
im_df = pd.DataFrame.from_dict(im_dict, orient = 'index')
#add column name
im_df.columns = ['infant_mortality']

#sort by infant mortality rate, sort_values ascending parameter defaults to true
im_df.sort_values('infant_mortality').head(10)
```

```python
#2 10 Cities with Highest Population

#initialize dictionary
pop_dict = {}

#start looping
for country in document.iterfind('country'):
    cities = country.findall('city')
    provinces = country.findall('province')
    #get the regular city populations
    for city in cities:
        city_name = city.find('name').text
        pops = city.findall('population[last()]')
        for pop in pops:
            pop_count = int(pop.text)
            pop_dict[city_name] = pop_count
    #get the province city populations        
    for province in provinces :
        p_cities= province.findall('city')
        for p_city in p_cities :
            p_city_name = p_city.find('name').text
            p_pops = p_city.findall('population[last()]') #we just want the latest population count
            for p_pop in p_pops :
                pop_count = int(p_pop.text)
                pop_dict[p_city_name] = pop_count
        
pop_df = pd.DataFrame.from_dict(pop_dict, orient = 'index')
pop_df.columns = ['population']
pop_df.sort_values('population', ascending=False).head(10)
```

```python
#3. 10 ethnic groups with the largest overall populations (sum of best/latest estimates over all countries)

#initialize dictionary
eth_dict = {}

#start looping
for country in document.iterfind('country'):
    ethnicities = country.findall('ethnicgroup')
    populations = country.findall('population[last()]') #we just want the latest population count
    for pop in populations:
        pop_count = float(pop.text)
    for ethnicity in ethnicities:
        ethnicity_name = ethnicity.text
        ethnicity_d = ethnicity.attrib
        ethnicity_tot = float(ethnicity_d['percentage']) * pop_count / 100.0 #pct of pop * pop = ethnic pop
        if ethnicity_name in eth_dict :
            eth_dict[ethnicity_name] += ethnicity_tot
        else :
            eth_dict[ethnicity_name] = ethnicity_tot


eth_df = pd.DataFrame.from_dict(eth_dict, orient = 'index')
eth_df.columns = ['population']
eth_df.sort_values('population', ascending = False).head(10)
```

```python
#4 name and country of a) longest river, b) largest lake and c) airport at highest elevation

#create dictionary of countries and codes
country_dict = {}

for country in document.iterfind('country'):
    name = country.find('name')
    country_dict[country.attrib['car_code']] = name.text

#river
river_dict = {}
temp_dict = {}

for river in document.iterfind('river'):
    name = river.find('name')
    country = str(river.attrib['country'])
    if ' ' in country : #lets assume first country in list is 'owning' country
        country = river.attrib['country'].partition(' ')[0] #grabs first piece of newly partitioned list
    length = river.find('length')
    if (length != None) : #replace car_code w/ ctry name as we're building dictionary
        temp_dict = {'length':float(length.text), 'country': country_dict[country]} 
        river_dict[name.text] = temp_dict
river_df = pd.DataFrame.from_dict(river_dict, orient = 'index')
river_df.sort_values('length', ascending = False).head(1)
```

```python
#lake
lake_dict = {}
temp_dict = {}

for lake in document.iterfind('lake'):
    name = lake.find('name')
    country = str(lake.attrib['country'])
    if ' ' in country : #lets assume first country in list is 'owning' country
        country = lake.attrib['country'].partition(' ')[0] #grabs first piece of newly partitioned list
    area = lake.find('area')
    if (area != None) : #replace car_code w/ ctry name as we're building dictionary
        temp_dict = {'area': float(area.text), 'country':country_dict[country]}
        lake_dict[name.text] = temp_dict
lake_df = pd.DataFrame.from_dict(lake_dict, orient = 'index')
lake_df.sort_values('area', ascending = False).head(1)   
```

```python
#airport
ap_dict = {}
temp_dict = {}

for airport in document.iterfind('airport') :
    name = airport.find('name')
    elevation = airport.find('elevation')
    if elevation.text != None : #replace ctry code w/ ctry name as we're building dictionary
        temp_dict = {'elevation':float(elevation.text), 'country':country_dict[airport.attrib['country']]}
        ap_dict[name.text] = temp_dict
        
ap_df = pd.DataFrame.from_dict(ap_dict, orient = 'index')
ap_df.sort_values('elevation', ascending = False).head(1)
```
